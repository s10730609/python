```python
Folder_Path = r'D:\Users\side project\dataset\archive\country_vaccinations.csv'
dfv = pd.read_csv(Folder_Path)
```

```python
dfv.head()
```

```python
all_vaccine = []
#把原dataset vaccine切割
for x in dfv.vaccines:
    all_vaccine.extend(x.split(', '))
#把原dataset vaccine去除空白
# vaccines = [x.strip() for x in all_vaccine]
vaccines = pd.unique(all_vaccine)
vaccines
```

```python
zero_matrix = np.zeros((len(dfv), len(vaccines)))
dummies = pd.DataFrame(zero_matrix, columns=vaccines)
```

```python
dummies
```

```python
vac = dfv.vaccines[0]
vac.split(', ')
dummies.columns.get_indexer(vac.split(', '))
```

```python
for i, vac in enumerate(dfv.vaccines):
    indices = dummies.columns.get_indexer(vac.split(', '))
    dummies.iloc[i, indices] = 1
```

```python
dfv_2 = dfv.join(dummies.add_prefix('vaccine_'))
```

```python
dfv_2 = dfv_2.fillna(0)
dfv_2
```

```python
dfv_2.to_csv(r"D:\\Users\\side project\\dataset\\archive\\country_vaccinations_fin.csv", encoding="utf_8_sig", index=False, header=True)
```
